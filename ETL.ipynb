Generar campo id: Cada id se compondrá de la primera letra del nombre de la plataforma, seguido del show_id ya presente en los datasets (ejemplo para títulos de Amazon = as123)

```python
# generamos una copia para no alterar el dataset original 
Amazon_1 = Amazon.copy()
Disney_1 = Disney.copy()
Hulu_1 = Hulu.copy() 
Netflix_1 = Netflix.copy()
```

```python
#Primero voy a agregar a todos los datasts una columna llamada 'platform'
Amazon_1['platform'] = 'amazon'
Disney_1['platform'] = 'disney'
Hulu_1['platform'] = 'hulu'
Netflix_1['platform'] = 'netflix'
```

```python
# Generamos el campo id: Cada id se compondrá de la primera letra del
# nombre de la plataforma, seguido del show_id ya presente en los dataset

Amazon_1.insert(loc=0, column="Id", value="a" + Amazon["show_id"])
Disney_1.insert(loc=0, column="Id", value="d" + Disney["show_id"])
Hulu_1.insert(loc=0, column="Id", value="h" + Hulu["show_id"])
Netflix_1.insert(loc=0, column="Id", value="n" + Netflix["show_id"])
```

Los valores nulos del campo rating deberán reemplazarse por el string “G” (corresponde al maturity rating: “general for all audiences”

```python
# verificamos la cantidad de nulos por columna

peliculas_1.isnull().sum()
```

```python
# Reemplazamos los valores nulos del campo "Rating" por la letra "G"

peliculas_1["rating"].fillna("G", inplace=True)
```

```python
# Verificamos la correcciòn sobre la columna "rating"
peliculas_1.isnull().sum()
```

De haber fechas, deberán tener el formato AAAA-mm-dd

```python
# modificamos el formato de la columna "date_added" a AAAA-mm-dd
peliculas_1["date_added"] = pd.to_datetime(peliculas_1["date_added"])
```

```python
#visualizamos 
peliculas_1.head(2)
```

Los campos de texto deberán estar en minúsculas, sin excepciones

```python
# Pasamos a minúsculas los campos de texto
peliculas_1["type"] = peliculas_1["type"].str.lower()
peliculas_1["title"] = peliculas_1["title"].str.lower()
peliculas_1["director"] = peliculas_1["director"].str.lower()
peliculas_1["cast"] = peliculas_1["cast"].str.lower()
peliculas_1["country"] = peliculas_1["country"].str.lower()
peliculas_1["rating"] = peliculas_1["rating"].str.lower()
peliculas_1["duration"] = peliculas_1["duration"].str.lower()
peliculas_1["listed_in"] = peliculas_1["listed_in"].str.lower()
peliculas_1["description"] = peliculas_1["description"].str.lower()
```

```python
# visualizamos
peliculas_1.head(2)
```

El campo duration debe convertirse en dos campos: duration_int y duration_type. El primero será un integer y el segundo un string indicando la unidad de medición de duración: min (minutos) o season (temporadas)

```python
# Separamos el campo "duration" en 2 columnas
splitcolumnas = peliculas_1["duration"].str.split(" ", n=1, expand=True)
```

```python
# Insertamos una columna con los minutos o las temporadas 
peliculas_1["duration_int"] = splitcolumnas[0]
```

```python
# Insertamos una columna con las unidades 
peliculas_1["duration_type"] = splitcolumnas[1]
```

```python
# Reemplazamos "seasons" por "season" en la columna "duration_type"
peliculas_1["duration_type"] = peliculas_1["duration_type"].str.replace("seasons", "season")
```

```python
# Cambiamor el formato del campo "duration_int" a integer
peliculas_1["duration_int"] = peliculas_1["duration_int"].astype("Int64")
```

```python
# eliminamos la columna duration
peliculas_2= peliculas_1.drop(["duration"],axis=1)
```

teniendo el dataset, se utiliza los csvs de la carpeta ratings, donde los uniremos y luego juntaremos estos dos csv para tener un csv final completo 

```python
#Cargo los csv
rating_1 = pd.read_csv('ratings/1.csv', sep=",")
rating_2 = pd.read_csv('ratings/2.csv', sep=",")
rating_3 = pd.read_csv('ratings/3.csv', sep=",")
rating_4 = pd.read_csv('ratings/4.csv', sep=",")
rating_5 = pd.read_csv('ratings/5.csv', sep=",")
rating_6 = pd.read_csv('ratings/6.csv', sep=",")
rating_7 = pd.read_csv('ratings/7.csv', sep=",")
rating_8 = pd.read_csv('ratings\8.csv', sep=",")
```

```python
#unimos todo los dataset

ratings_unidos = pd.concat([rating_1, rating_2, rating_3, rating_4, rating_5, rating_6, rating_7, rating_8])
```

```python
#revisamos si estan correctamente  
ratings_unidos.head()
```

```python
#numero de filas y columnas 
ratings_unidos.shape
```

```python
#resuminos toda su información
ratings_unidos.info()
```

```python
#cambio el formato de la columna 'timestamp', aca lo paso de int a 'float'
ratings_unidos['timestamp'] = pd.to_datetime(ratings_unidos['timestamp'], unit='s').dt.strftime('%B %d, %Y')
```

```python
#paso a formato fecha 
ratings_unidos['timestamp'] = pd.to_datetime(ratings_unidos['timestamp'])
```

```python
#Cambiamos el nombre del campo 'rating' por 'score'
ratings_unidos2 = ratings_unidos.rename(columns = {'rating' : 'score'})
```

```python
#valido que haya quedado el cambio
ratings_unidos2
```

exportamos el rating unificado con las transformaciones 

```python
ratings_unidos2.to_csv("MoviesScore.csv", index=False)
```

```python
#Calculo el promedio de cada pelicula 
MeanScore = ratings_unidos2.groupby("movieId")
MeanScore = MeanScore["score"].mean()
```

```python
#Lo pasamos a un dataframe y reseteamos el indice
df_MeanScore= pd.DataFrame(MeanScore)
```

```python
#revisamos los cambios
df_MeanScore
```

```python
platform_movies = peliculas_2.merge(df_MeanScore, left_on="Id", right_on="movieId")
```
